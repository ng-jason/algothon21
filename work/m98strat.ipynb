For sideways trading stocks find convergence

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from numpy.polynomial.polynomial import polyfit
data = np.loadtxt("prices250.txt")
data = np.transpose(data)
```

for stock #55

```python
plt.plot(data[55])
plt.show()
```

```python
SEQUENCE_LENGTH = 30
SEQUENCES = 69
seq = []
for i in range (0, SEQUENCES):
    seq_temp = []
    for j in range(0, SEQUENCE_LENGTH):
        seq_temp.append(data[55][np.random.randint(0, 150)])
    seq.append(seq_temp)
    
seq_vector = []
for j in range(1, SEQUENCE_LENGTH+1):
    seq_vector.append(j)   
    
convergence_points = []
for s in seq:
    x = np.array(seq_vector)
    y = np.array(s)
    m, b = polyfit(x, y, 1)
    pred = 250*(b) + m
    convergence_points.append(pred)
    
convg = 0
for c in convergence_points:
    convg += c
convg = convg / len(convergence_points)
```

```python
x_values = [0, 250]
y_values = [data[55][0], convg]
plt.plot(x_values, y_values)
plt.plot(data[55][:150])
plt.show()
b2, m2 = np.polyfit(x_values, y_values, 1)
```

```python
# get convergence line for 150 days
def get_line(stock_num):
    SEQUENCE_LENGTH = 30
    SEQUENCES = 69
    seq = []
    for i in range (0, SEQUENCES):
        seq_temp = []
        for j in range(0, SEQUENCE_LENGTH):
            seq_temp.append(data[stock_num][np.random.randint(0, 150)])
        seq.append(seq_temp)
    seq_vector = []
    for j in range(1, SEQUENCE_LENGTH+1):
        seq_vector.append(j)   
    convergence_points = []
    for s in seq:
        x = np.array(seq_vector)
        y = np.array(s)
        m, b = polyfit(x, y, 1)
        pred = 250*(b) + m
        convergence_points.append(pred)
    convg = 0
    for c in convergence_points:
        convg += c
    convg = convg / len(convergence_points)
    x_values = [0, 250]
    y_values = [data[stock_num][0], convg]
    b2, m2 = np.polyfit(x_values, y_values, 1)
    print(b2, m2)
```
