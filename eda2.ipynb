```python
window = 20
data['min'] = data[0].rolling(window=window).min()
data['mean'] = data[0].rolling(window=window).mean()
data['std'] = data[0].rolling(window=window).std()
data['median'] = data[0].rolling(window=window).median()
data['max'] = data[0].rolling(window=window).max()
data['ewma'] = data[0].ewm(halflife=0.5, min_periods=window).mean()
```

```python
# plot three rolling statistics
ax = data[['min', 'mean', 'max']].iloc[:250].plot(figsize=(10, 8))  
data[0].iloc[:250].plot(ax=ax) # adds time series data to plot
```

```python
# plot price for instrument0 plus two SMA time series
data['SMA01'] = data[0].rolling(window=5).mean()  # calculates simple moving average
data['SMA02'] = data[0].rolling(window=50).mean()
data[[0, 'SMA01', 'SMA02']].plot(figsize=(10, 8))
```

```python
# the purple is +1 when you should buy or hold and -1 when you should sell
# when moving averages cross over

data['positions0'] = np.where(data['SMA01'] > data['SMA02'], 1, -1)
ax = data[[0, 'SMA01', 'SMA02', 'positions0']].plot(figsize=(10, 8), secondary_y = 'positions0')
ax.get_legend().set_bbox_to_anchor((0.25, 0.85))
```

```python
rets = np.log(data / data.shift(1))  # 
rets.dropna(inplace=True)
rets.iloc[:, :2].plot(subplots=True, figsize=(10, 8))
```

```python
n_instruments = 2
# plots scatter matrix of first n instruments
pd.plotting.scatter_matrix(rets.iloc[:, :n_instruments],
                           alpha=0.8,
                           diagonal='hist',
                           hist_kwds={'bins':35},
                           figsize=(10, 10))
```
