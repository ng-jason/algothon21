```python
# from https://www.learndatasci.com/tutorials/python-finance-part-yahoo-finance-api-pandas-matplotlib
zero = df.loc[:, 0]

# Calculate the 20 and 100 days moving averages of the closing prices
short_rolling_zero = zero.rolling(window=20).mean()
long_rolling_zero = zero.rolling(window=100).mean()

# Plot everything by leveraging the very powerful matplotlib package
fig, ax = plt.subplots(figsize=(16,9))

ax.plot(zero.index, zero, label='0')
ax.plot(short_rolling_zero.index, short_rolling_zero, label='20 days rolling')
ax.plot(long_rolling_zero.index, long_rolling_zero, label='100 days rolling')

ax.set_xlabel('Day')
ax.set_ylabel('Price ($)')
ax.legend()
```
